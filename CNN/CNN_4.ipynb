4: Face Detection from an image using OpenCV

```python
import urllib.request

url = "https://raw.githubusercontent.com/opencv/opencv/master/data/haarcascades/haarcascade_frontalface_default.xml"
filename = "haarcascade_frontalface_default.xml"
urllib.request.urlretrieve(url, filename)
```

```python
# Load the Haar cascade classifier
import cv2
face_cascade = cv2.CascadeClassifier("haarcascade_frontalface_default.xml")
```

```python
# Load the image
image_path = '/Users/rahulkp/Downloads/istockphoto-1146473249-612x612.jpg'  
image = cv2.imread(image_path)
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
# Detect faces in the image
faces = face_cascade.detectMultiScale(gray, scaleFactor=1.2, minNeighbors=5, minSize=(30, 30))
```

```python
# Draw rectangles around the detected faces
for (x, y, w, h) in faces:
    cv2.rectangle(image, (x, y), (x + w, y + h), (255, 0, 0), 2)
```

```python
# Convert image from BGR to RGB for matplotlib display
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
# Display the image with detected faces using matplotlib
import matplotlib.pyplot as plt
plt.imshow(image_rgb)
plt.axis('off')
plt.show()
```
